The first two PCs capture most of the information from the original dataset so they have a clear trend of dropping in values.

````python
### YOUR CODE HERE ###
import numpy as np
data_values = cases_std.values
cov_matrix = np.cov(data_values.T)
assert cov_matrix.shape == (483, 483), "ERROR: cov matrix shape mismatch."

# get eigenvalues/vectors
def get_sorted_eigen(df_cov):
  eigenValues, eigenVectors = np.linalg.eigh(df_cov) 
  return eigenValues, eigenVectors

eigenValues, eigenVectors = get_sorted_eigen(cov_matrix)

# show scree plot
  #sort in decreasing orders
args = (-eigenValues).argsort()
eigenValues = eigenValues[args]
eigenVectors = eigenVectors[:, args]
  # make plot
eigValSum = sum(eigenValues)
expVar = [eigV/eigValSum*100 for eigV in eigenValues]
cumExpVar = np.cumsum(expVar)
plt.bar(range(483), expVar, label='Explained Variance')
plt.plot(cumExpVar, 'r-o', label='Cumulative Explained Variance')
plt.legend()
plt.show()
````

````python
plt.bar(range(20), expVar[0:20], label='Explained Variance')
plt.plot(cumExpVar[0:20], 'r-o', label='Cumulative Explained Variance')
plt.legend()
plt.show()
````

````python
# find #of PCs needed to cover 99% of variance
for i in range(20):
  if cumExpVar[i] >= 99:
    num_pc = i
    break
  else:
    pass

print('The number of PCs needed to cover >= 99% variance: ', num_pc +1)
````

````python
# Plot the first 16 principal components (Eigenvectors) as a time series
eigenVectors_16 = eigenVectors[:, 0:16].T
assert eigenVectors_16.shape == (16, 483), "ERROR: 16 eigenvectors not correct"

time_list = range(1, 484) # x-axis is date
for i in range (eigenVectors_16.shape[0]):
  plt.plot(time_list, eigenVectors_16[i])
  plt.xlabel('Time')
  plt.ylabel('PC_value')
  plt.title(f"PC{i+1}")
  plt.show()
  i += 1
````

````python
from sklearn.metrics import mean_squared_error

def plot_country_figures(original_df, country_name):
  #original time-series
  cases_raw.loc[country_name].plot()
  plt.xlabel('Date')
  plt.ylabel('# vaccines')
  plt.title(f'Original time series: {country_name}')
  plt.show()
  #The incremental reconstruction of the original (not standardized) time-series
  num_pcs = [1,2,4,8,16]
  reconstructions =[]
  for num in num_pcs:
    eigenvector_select = eigenVectors[:, 0:num] # select PC
    assert eigenvector_select.shape == (483, num), "Wrong eigenvector shape"
    projection = np.dot(cases_std.loc[country_name], eigenvector_select) #project on selected PCs
    ReconX = np.dot(projection, eigenvector_select.T) #reconstruction
    scaler = StandardScaler()
    scaler.fit_transform(original_df) # first fit and then inverse it
    Reconx_origin = scaler.inverse_transform(ReconX.reshape(1,-1))
    assert Reconx_origin.shape == (1,483), "reconstructed original data shape mismatch!"
    Reconx_origin = Reconx_origin.reshape(483) # reshape for plotting

    reconstructions.append(Reconx_origin)
  # plot the time-series
  i = 0
  for num in num_pcs:
    reconstruction = reconstructions[i]
    plt.plot(reconstruction, label=f'first {num} PCs')
    plt.title(f'Incremental reconstruction of original time series: {country_name}')
    plt.xlabel('Date')
    plt.ylabel('# vaccines')
    plt.legend()
    i += 1
  plt.show()

  # find the residual error for your best reconstruction (PC16) with respect to the original time-series.
  Reconx_origin_16 = reconstructions[-1] 
  residual = original_df.loc[country_name].values - Reconx_origin_16
  plt.plot(residual)
  plt.title(f'Residual error with respect to original (16 PCs): {country_name}')
  plt.xlabel('Date')
  plt.ylabel("Residual")
  plt.show()

  #The RMSE of the reconstruction as a function of the number of included components
  reconstructions_RMSE = []
  for num in range(1,11):
    eigenvector_select = eigenVectors[:, 0:num] # select PC
    assert eigenvector_select.shape == (483, num), "Wrong eigenvector shape"
    projection = np.dot(cases_std.loc[country_name], eigenvector_select) #project on selected PCs
    ReconX = np.dot(projection, eigenvector_select.T) #reconstruction
    scaler = StandardScaler()
    scaler.fit_transform(original_df) # first fit and then inverse it
    Reconx_origin = scaler.inverse_transform(ReconX.reshape(1,-1))
    assert Reconx_origin.shape == (1,483), "reconstructed original data shape mismatch!"
    Reconx_origin = Reconx_origin.reshape(483) # reshape for plotting

    reconstructions_RMSE.append(Reconx_origin)
  assert len(reconstructions_RMSE) == 10, "reconstructions_RMSE length mismatch!"
  
  rmse_list = []
  for item in reconstructions_RMSE:
    rmse = mean_squared_error(original_df.loc[country_name].values, item) ** 0.5 # calculate RMSE
    rmse_list.append(rmse)
  assert len(rmse_list) == 10, "RMSE length mismatch!"
  plt.plot(rmse_list)
  plt.title(f'RMSE of reconstruction: {country_name}')
  plt.xlabel('Number of PCs')
  plt.ylabel("RMSE")
  plt.show()


````

````python
### YOUR CODE HERE ###
from sklearn.decomposition import TruncatedSVD 
def plot_country_figures_svd(original_df, country_name):
  #original time-series
  cases_raw.loc[country_name].plot()
  plt.xlabel('Date')
  plt.ylabel('# vaccines')
  plt.title(f'Original time series: {country_name}')
  plt.show()
  #The incremental reconstruction of the original (not standardized) time-series
  num_pcs = [1,2,4,8,16]
  reconstructions =[]
  for num in num_pcs:
    # use SVD
    scaler = StandardScaler()
    scaler.fit_transform(original_df) # fit scaler

    svd = TruncatedSVD(n_components=num)
    svd.fit_transform(cases_std.values) # fit svd on standardized data
    projection = svd.transform(cases_std.loc[country_name].values.reshape(1,-1))
    Reconx = svd.inverse_transform(projection)
    Reconx_origin = scaler.inverse_transform(Reconx.reshape(1,-1))
    Reconx_origin = Reconx_origin.reshape(483) # reshape for plotting
    reconstructions.append(Reconx_origin)
  # plot the time-series
  i = 0
  for num in num_pcs:
    reconstruction = reconstructions[i]
    plt.plot(reconstruction, label=f'first {num} SVD components')
    plt.title(f'Incremental reconstruction of original time series: {country_name}')
    plt.xlabel('Date')
    plt.ylabel('# vaccines')
    plt.legend()
    i += 1
  plt.show()

  # find the residual error for your best reconstruction (16 components) with respect to the original time-series.
  Reconx_origin_16 = reconstructions[-1] 
  residual = original_df.loc[country_name].values - Reconx_origin_16
  plt.plot(residual)
  plt.title(f'Residual error with respect to original (16 SVD componets): {country_name}')
  plt.xlabel('Date')
  plt.ylabel("Residual")
  plt.show()

  #The RMSE of the reconstruction as a function of the number of included components
  reconstructions_RMSE = []
  for num in range(1,11):
    # use SVD
    scaler = StandardScaler()
    scaler.fit_transform(original_df) # fit scaler

    svd = TruncatedSVD(n_components=num)
    svd.fit_transform(cases_std.values) # fit svd on standardized data
    projection = svd.transform(cases_std.loc[country_name].values.reshape(1,-1))
    Reconx = svd.inverse_transform(projection)
    Reconx_origin = scaler.inverse_transform(Reconx.reshape(1,-1))
    Reconx_origin = Reconx_origin.reshape(483) # reshape for plotting

    reconstructions_RMSE.append(Reconx_origin)
  assert len(reconstructions_RMSE) == 10, "reconstructions_RMSE length mismatch!"
  
  rmse_list = []
  for item in reconstructions_RMSE:
    rmse = mean_squared_error(original_df.loc[country_name].values, item) ** 0.5 # calculate RMSE
    rmse_list.append(rmse)
  assert len(rmse_list) == 10, "RMSE length mismatch!"
  plt.plot(rmse_list)
  plt.title(f'RMSE of reconstruction: {country_name}')
  plt.xlabel('Number of SVD components')
  plt.ylabel("RMSE")
  plt.show()
````

````python
def plot_country_figures_svd_part5(original_df, country_name):
  #original time-series
  deaths_raw.loc[country_name].plot()
  plt.xlabel('Date')
  plt.ylabel('# deaths')
  plt.title(f'Original time series: {country_name}')
  plt.show()
  #The incremental reconstruction of the original (not standardized) time-series
  num_pcs = [1,2,4,8,16]
  reconstructions =[]
  for num in num_pcs:
    # use SVD
    scaler = StandardScaler()
    scaler.fit_transform(original_df) # fit scaler

    svd = TruncatedSVD(n_components=num)
    svd.fit_transform(deaths_std.values) # fit svd on standardized data
    projection = svd.transform(deaths_std.loc[country_name].values.reshape(1,-1))
    Reconx = svd.inverse_transform(projection)
    Reconx_origin = scaler.inverse_transform(Reconx.reshape(1,-1))
    Reconx_origin = Reconx_origin.reshape(733) # reshape for plotting
    reconstructions.append(Reconx_origin)
  # plot the time-series
  i = 0
  for num in num_pcs:
    reconstruction = reconstructions[i]
    plt.plot(reconstruction, label=f'first {num} SVD components')
    plt.title(f'Incremental reconstruction of original time series: {country_name}')
    plt.xlabel('Date')
    plt.ylabel('# deaths')
    plt.legend()
    i += 1
  plt.show()

  # find the residual error for your best reconstruction (16) with respect to the original time-series.
  Reconx_origin_16 = reconstructions[-1] 
  residual = original_df.loc[country_name].values - Reconx_origin_16
  plt.plot(residual)
  plt.title(f'Residual error with respect to original (16 SVD components): {country_name}')
  plt.xlabel('Date')
  plt.ylabel("Residual")
  plt.show()

  #The RMSE of the reconstruction as a function of the number of included components
  reconstructions_RMSE = []
  for num in range(1,11):
    # use SVD
    scaler = StandardScaler()
    scaler.fit_transform(original_df) # fit scaler

    svd = TruncatedSVD(n_components=num)
    svd.fit_transform(deaths_std.values) # fit svd on standardized data
    projection = svd.transform(deaths_std.loc[country_name].values.reshape(1,-1))
    Reconx = svd.inverse_transform(projection)
    Reconx_origin = scaler.inverse_transform(Reconx.reshape(1,-1))
    Reconx_origin = Reconx_origin.reshape(733) # reshape for plotting

    reconstructions_RMSE.append(Reconx_origin)
  assert len(reconstructions_RMSE) == 10, "reconstructions_RMSE length mismatch!"
  
  rmse_list = []
  for item in reconstructions_RMSE:
    rmse = mean_squared_error(original_df.loc[country_name].values, item) ** 0.5 # calculate RMSE
    rmse_list.append(rmse)
  assert len(rmse_list) == 10, "RMSE length mismatch!"
  plt.plot(rmse_list)
  plt.title(f'RMSE of reconstruction: {country_name}')
  plt.xlabel('Number of SVD components')
  plt.ylabel("RMSE")
  plt.show()
````

````python
# compare the results of PCA and SVD

def compare_PCA_SVD(countries, df_original, df_std): # compare the RMSE for PCA and SVD
  for country in countries:
    # SVD
    reconstructions_RMSE = []
    for num in range(1,11):
      scaler = StandardScaler()
      scaler.fit_transform(df_original) # fit scaler

      svd = TruncatedSVD(n_components=num)
      svd.fit_transform(df_std.values) # fit svd on standardized data
      projection = svd.transform(df_std.loc[country].values.reshape(1,-1))
      Reconx = svd.inverse_transform(projection)
      Reconx_origin = scaler.inverse_transform(Reconx.reshape(1,-1))
      Reconx_origin = Reconx_origin.reshape(733) # reshape for plotting

      reconstructions_RMSE.append(Reconx_origin)
    assert len(reconstructions_RMSE) == 10, "reconstructions_RMSE length mismatch!"
  
    rmse_list = []
    for item in reconstructions_RMSE:
      rmse = mean_squared_error(df_original.loc[country].values, item) ** 0.5 # calculate RMSE
      rmse_list.append(rmse)
    assert len(rmse_list) == 10, "RMSE length mismatch!"
    plt.plot(rmse_list, label = 'SVD')
    plt.title(f'RMSE of reconstruction: {country}')
    plt.xlabel('Number of components')
    plt.ylabel("RMSE")
    plt.legend()
    # PCA
    reconstructions_RMSE = []
    for num in range(1,11):
      eigenvector_select = eigenVectors[:, 0:num] # select PC
      assert eigenvector_select.shape == (733, num), "Wrong eigenvector shape"
      projection = np.dot(df_std.loc[country], eigenvector_select) #project on selected PCs
      ReconX = np.dot(projection, eigenvector_select.T) #reconstruction
      scaler = StandardScaler()
      scaler.fit_transform(df_original) # first fit and then inverse it
      Reconx_origin = scaler.inverse_transform(ReconX.reshape(1,-1))
      assert Reconx_origin.shape == (1,733), "reconstructed original data shape mismatch!"
      Reconx_origin = Reconx_origin.reshape(733) # reshape for plotting

      reconstructions_RMSE.append(Reconx_origin)
    assert len(reconstructions_RMSE) == 10, "reconstructions_RMSE length mismatch!"
  
    rmse_list = []
    for item in reconstructions_RMSE:
      rmse = mean_squared_error(df_original.loc[country].values, item) ** 0.5 # calculate RMSE
      rmse_list.append(rmse)
    assert len(rmse_list) == 10, "RMSE length mismatch!"
    plt.plot(rmse_list,label = 'PCA')
    plt.title(f'RMSE of reconstruction: {country}')
    plt.xlabel('Number of components')
    plt.ylabel("RMSE")
    plt.legend()
    plt.show()


compare_PCA_SVD(countries, deaths_raw, deaths_std)
````
